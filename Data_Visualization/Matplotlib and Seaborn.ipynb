### Line Chart

```python
x=[1,2,3,4,5,6,7,]
y=[50,60,65,40,55,70,45]
```

```python
plt.plot(x,y)
```

```python

plt.plot(x,y,color='green', linewidth=3)
```

```python
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(x,y,color='green', linewidth=3)
```

```python
days=[1,2,3,4,5,6,7]
max=[50,51,52,48,47,49,46]
min=[43,42,40,44,33,35,37]
avg=[45,48,40,46,40,42,41]
```

```python
plt.plot(days,max)
plt.plot(days,min)
plt.plot(days,avg)
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days,max,label='max')
plt.plot(days,min,label='min')
plt.plot(days,avg,label='avg')
plt.legend()
#plt.legend('upper right')
#plt.grid()
```
